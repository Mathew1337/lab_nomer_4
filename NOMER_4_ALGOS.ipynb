```
import random
import string

class URLShortener:
    def __init__(self, min_length=4, max_length=8):
        self.url_map = {}
        self.reverse_map = {}
        self.click_count = {}
        self.min_length = min_length
        self.max_length = max_length

    def _generate_code(self, length):
        return ''.join(random.choice(string.ascii_letters + string.digits) for _ in range(length))

    def add_url(self, long_url):
        if not (long_url.startswith('http://') or long_url.startswith('https://')):
            raise ValueError("Ссылка должна начинаться с http:// или https://")

        if long_url in self.reverse_map:
            return self.reverse_map[long_url]

        while True:
            length = random.randint(self.min_length, self.max_length)
            short_code = self._generate_code(length)
            if short_code not in self.url_map:
                break

        self.url_map[short_code] = long_url
        self.reverse_map[long_url] = short_code
        self.click_count[short_code] = 0
        return short_code

    def get_url(self, short_code):
        if short_code in self.url_map:
            self.click_count[short_code] += 1
            return self.url_map[short_code]
        return None

    def exists(self, short_code):
        return short_code in self.url_map

    def get_all_links(self):
        return list(self.url_map.items())

    def delete_url(self, short_code):
        if short_code in self.url_map:
            long_url = self.url_map[short_code]
            del self.url_map[short_code]
            del self.reverse_map[long_url]
            del self.click_count[short_code]
            return True
        return False

    def get_popular(self):
        if not self.click_count:
            return None
        return max(self.click_count.items(), key=lambda x: x[1])

    def search_by_word(self, word):
        return [(code, url) for code, url in self.url_map.items() if word.lower() in url.lower()]

    def get_sorted_by_code(self):
        return sorted(self.url_map.items(), key=lambda x: x[0])

    def update_url(self, short_code, new_long_url):
        if not (new_long_url.startswith('http://') or new_long_url.startswith('https://')):
            raise ValueError("Ссылка должна начинаться с http:// или https://")

        if short_code in self.url_map:
            old_url = self.url_map[short_code]
            del self.reverse_map[old_url]
            self.url_map[short_code] = new_long_url
            self.reverse_map[new_long_url] = short_code
            return True
        return False

    def find_by_long_url(self, long_url):
        return self.reverse_map.get(long_url)


if __name__ == "__main__":
    s = URLShortener(min_length=4, max_length=6)

    c1 = s.add_url("https://google.com")
    c2 = s.add_url("https://github.com")
    c3 = s.add_url("https://youtube.com")
    c4 = s.add_url("https://google.com")

    print(f"Добавлено: {c1}, {c2}, {c3}")
    print(f"Дубликат вернул: {c4}")

    print(f"\nget_url({c1}): {s.get_url(c1)}")
    print(f"exists({c1}): {s.exists(c1)}")
    print(f"exists(abc): {s.exists('abc')}")

    print(f"\nВсе ссылки: {s.get_all_links()}")

    s.get_url(c1)
    s.get_url(c1)
    print(f"\nПереходов по {c1}: {s.click_count[c1]}")

    print(f"Самая популярная: {s.get_popular()}")

    print(f"Поиск 'google': {s.search_by_word('google')}")

    print(f"Сортировка по коду: {s.get_sorted_by_code()}")

    s.update_url(c2, "https://github.com/python")
    print(f"После обновления {c2}: {s.get_url(c2)}")

    print(f"Обратный поиск google: {s.find_by_long_url('https://google.com')}")

    s.delete_url(c3)
    print(f"После удаления {c3}: {s.get_all_links()}")
```
